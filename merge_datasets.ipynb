```python
import subprocess
import os

## define your paths
path1 = 'custom_data/Custom_Dataset_Kyler/'
path2 = 'custom_data/George_Custom_Dataset/'

## where to place the merged data
merged_path = 'custom_data/merged/'

subs = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 
        'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

for sub in subs:
    for opt in ['left', 'right']:
    
        path1_use = path1 + sub + '/' + opt + '/'
        path2_use = path2+ sub + '/' + opt + '/'
        merged_path_use = merged_path + sub + '/' + opt + '/'
        
        if not os.path.exists(merged_path + sub + '/'):
            os.mkdir(merged_path + sub + '/')
        
        if not os.path.exists(merged_path_use): # Create the subdirectories if dont already exist
            os.mkdir(merged_path_use)

        ## write an rsync commands to merge the directories
        rsync_cmd = 'rsync' + ' -avzh ' + path1_use + ' ' + path2_use + ' ' + merged_path_use
        ## run the rsync command
        subprocess.run(rsync_cmd, shell=True)
```
